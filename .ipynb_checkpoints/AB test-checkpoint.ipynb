# Introduction

The company is looking to improve sales by changing the webpage design. They are conducting an AB test to see if new page design has any significant effect on sales. Two equally-sized groups are created - Treatment and Control groups. Data is cleaned and periodicly monitored through out the experiment.

        </ul></p>
     <p><li><b>Hypothesis Testing</b></li>

### About Dataset

<b>user_id:</b> unique users number

<b>timestamp:</b> time

<b>group:</b> treatment and control group

<b>landing_page:</b> old_page and new_page

<b>converted:</b> Users that bought something after viewing the page (0-1)

### AB Testing

#### Create Hypothesis

H0: There is not statistically significant difference between the old page and new page

H1: There is statistically significant difference between the old page and new page

### Experiment Definitions

<ul>
  <li><b>Control</b>: Getting old page</li>
  <li><b>Treatment</b>: Getting new page</li>
  <li><b>Bucket:</b>: Time slicer</li>
</ul>

As it is only a small amount of the data, we can just remove it.

I would like to simulate the process of the data gathering during the experiment so I will need to add a new column to dataframe which would put data into 4 buckets with iteration of 15 minutes.

But first we would need to convert 'timestamp' column to datetime format.

```python
dfu['timestamp'] = pd.to_datetime(dfu['timestamp'], format='%M:%S.%f')
dfu
```

Lets put conditions of iteration into variable

```python
conditions = [
    ((dfu['timestamp'] >= pd.to_datetime('1900-01-01 00:00:00.000')) & (dfu['timestamp'] <= pd.to_datetime('1900-01-01 00:15:00.000'))),
    ((dfu['timestamp'] > pd.to_datetime('1900-01-01 00:15:00.000')) & (dfu['timestamp'] <= pd.to_datetime('1900-01-01 00:30:00.000'))),
    ((dfu['timestamp'] > pd.to_datetime('1900-01-01 00:30:00.000')) & (dfu['timestamp'] <= pd.to_datetime('1900-01-01 00:45:00.000'))),
    ((dfu['timestamp'] > pd.to_datetime('1900-01-01 00:45:00.000')) & (dfu['timestamp'] <= pd.to_datetime('1900-01-01 00:59:59.900')))
    ]
```

Lets assign a value to our buckets or in other word groups

```python
values = [1, 2, 3 ,4 ]
```

And lets put it in np.select formula to asign the values

```python
dfu['Bucket'] = np.select(conditions, values)
dfu
```

Lets double check if all data is sorted into Buckets

```python
dfu['Bucket'].value_counts()
```

Now simulation can begin by Bucket

```python
time = 4 # Vary number to get experiment data by created Groups.
experiment_data = dfu[(dfu['Bucket'] >= 1) & (dfu['Bucket'] <= time)]
control = experiment_data[experiment_data['group']=='control']
treatment = experiment_data[experiment_data['group']=='treatment']

control_conversion_perc = round((control['converted'].sum() * 100)/ control['converted'].count(), 3)
treatment_conversion_perc = round((treatment['converted'].sum() * 100)/ treatment['converted'].count(), 3)
lift = round((treatment_conversion_perc - control_conversion_perc) / control_conversion_perc , 3)

print(f"Treatment Conversion Rate: {treatment_conversion_perc}%")
print(f"Control Conversion Rate: {control_conversion_perc}%")
print(f"Lift = {lift}%")
```

In order to determen which tests should be used in comparing the old page and the new one lets see data's variance homogeneity and normality.

### Assumption checks

Normality assumption

<ul>
   <li>H0: The assumption of normal distribution is provided</li>
   <li>H1: The assumption of normal distribution is not provided</li>
</ul>

If the p-value is less than 0.05, it is considered significant and a non-parametric test (mannwhitneyu test) will be used. Else a parametric test ( unpaired t-test)

```python
Image(filename='C:/Users/Ramiu/Desktop/parametric3.png')
```

```python
test_stat, pvalue = shapiro(dfu.loc[dfu["landing_page"] == "old_page", "converted"])
print("p-value:",pvalue)
print("test_stat:",test_stat)
```

```python
test_stat, pvalue = shapiro(dfu.loc[dfu["landing_page"] == "new_page", "converted"])
print("p-value:",pvalue)
print("test_stat:",test_stat)
```

p-value < 0.05, so assumption of normality is not provided. we will use non-parametric test(Mannâ€“Whitney U test)

Also, Q-Q plot can be used to see if data is folowing normal distribution.

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
fig = sm.qqplot(dfu.loc[dfu["landing_page"] == "old_page", "converted"], line = '45')
plt.show
```

```python
fig = sm.qqplot(dfu.loc[dfu["landing_page"] == "new_page", "converted"], line = '45')
plt.show
```

As the data is not alongside 45 degree line of Q-Q plot we can say that the assumption of normality is not provided.

### Variance Homogeneity

<ul>
   <li>H0: Variances are homogeneous</li>
   <li>H1: Variances are not homogeneous</li>
</ul>

```python
test_stat, pvalue = levene(dfu.loc[dfu["landing_page"] == "new_page", "converted"],
                           dfu.loc[dfu["landing_page"] == "old_page", "converted"])
print("p-value:",pvalue)  
print("test_stat:",test_stat)
```

Variances are homogeneous, because we can not recject the H0 hypothesis.

### Hypothesis Testing

a non-parametric, Mannâ€“Whitney U test
<ul>
   <li>0: There is not statistically significant difference between the old page and new page</li>
   <li>H1: There is statistically significant difference between the old page and new page</li>
</ul>

```python
test_stat, pvalue = mannwhitneyu(dfu.loc[df["landing_page"] == "new_page", "converted"],
                                 dfu.loc[df["landing_page"] == "old_page", "converted"])

print('Test Stat = %.4f, p-value = %.4f' % (test_stat, pvalue))
```

### Conclusion

p-value(0.1161) >0.05

We fail to reject zero. there is no statistically significant difference between the new page and the old page, so it does not bring a profit
